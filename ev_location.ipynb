```python
df1['Total Electric Vehicle']=df1['Total Electric Vehicle'].str.replace(',','').str.strip().astype(int)
```

```python
df1['Total Non-Electric Vehicle']=df1['Total Non-Electric Vehicle'].str.replace(',','').str.strip().astype(int)
```

```python
df1['Total']=df1['Total'].str.replace(',','').str.strip().astype(int)
```

```python
df1['percentage'] = df1['Total Electric Vehicle'] / df1['Total']
```

```python
high_ev_percentage = df1[df1['percentage'] >= avg]
low_ev_percentage = df1[df1['percentage'] < avg]
```

```python
d=low_ev_percentage.sort_values(by=['percentage'], ascending=False)[:7]
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Plotting Total Electric Vehicle on the first axis
ax1.bar(df1['State Name'], df1['Total Electric Vehicle'], label='Total Electric Vehicle', color='green')
ax1.set_ylabel('Number of Electric Vehicles')
ax1.set_yscale('log')
ax1.legend()

# Plotting Total Non-Electric Vehicle on the second axis
ax2.bar(df1['State Name'], df1['Total Non-Electric Vehicle'], label='Total Non-Electric Vehicle', color='orange')
ax2.set_ylabel('Number of Non-Electric Vehicles')
ax2.set_yscale('log')
ax2.legend()

plt.xticks(rotation=90)
plt.xlabel('State Name')
plt.title('Total Electric and Non-Electric Vehicles for Each State')
plt.tight_layout()
plt.show()
```

States with Lower EV Usage but Higher Non-EV Usage:

Mizoram has relatively lower electric vehicle usage compared to some other states but has a higher number of non-electric vehicles. Similarly, Arunachal Pradesh and Meghalaya also have lower EV usage but a relatively higher number of non-electric vehicles.

States with Higher EV Usage and Lower Non-EV Usage:

Delhi has relatively higher electric vehicle usage compared to some states and lower non-electric vehicle usage. Tripura, Uttar Pradesh, and Karnataka also have higher EV usage and lower non-electric vehicle usage. States with Balanced EV and Non-EV Usage:

Some states like Assam, Gujarat, Haryana, Kerala, Odisha, Punjab, and West Bengal have a significant number of electric vehicles and a considerable number of non-electric vehicles. States with Missing Data:

Some states like Andra Pradesh, Lakshadweep, Madhya Pradesh, and Telangana have missing data for both electric and non-electric vehicles.
